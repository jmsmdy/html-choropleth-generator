```python
import geopandas as gpd
import pandas as pd

US = gpd.read_file('US States/states.shp')
US.head()
```

```python
US['geometry_simplified'] = US['geometry'].apply(lambda x: x.simplify(0.05))
US['label'] = US['iso3166_2'].apply(lambda x: x.split('-')[1])
US['link'] = US['wikimedia'].apply(lambda x: f"https://en.wikipedia.org/wiki/{x[3:].replace(' ', '_')}")
```

```python
contiguous = list(pd.read_csv('contiguous.csv', header=None)[0])
contiguous[:5]
```

```python
US_contiguous = US[US['name'].isin(contiguous)]
US_contiguous.head()
```

```python
from IPython.core.display import display, HTML
from conversion import to_html

html = to_html(US_contiguous, 'label', 'link', 'geometry_simplified', 1000)

with open("output.html", "w") as f: 
    f.write(html) 

display(HTML(html))
```
